```python
brand_price_m = df.groupby('Brand')['Price'].mean().sort_values(ascending=False).plot(kind = "bar")

plt.title('Average Prices by Brands')
plt.ylabel('Average Prices ($)')
plt.ylabel('Average Prices ($)')
plt.xlabel('Brands')

for i in brand_price_m.patches:
    brand_price_m.annotate(f'{i.get_height():.2f}', 
                (i.get_x() + i.get_width() / 2, i.get_height()), 
                xytext=(0, 0), 
                textcoords='offset points', 
                ha='center', va='bottom')
plt.show()
```

```python
y = df ["Price"]
x = df[["Processor_Speed","RAM_Size", "Storage_Capacity"]]
```

```python
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(X_test)
```

```python
len(y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel = RandomForestRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_gridrfr = {"max_depth":[2,5,10,15,20,25],
                 "max_features" : ["auto","log2"],
                 "n_estimators" : [2,4,6,8,10,12]}
```

```python
gridrfr = GridSearchCV(rfmodel,param_gridrfr)
```

```python
gridrfr.fit(X_train,y_train)
```

```python
gridrfr.best_params_
```

```python
predictions =gridrfr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, predictions)
```
